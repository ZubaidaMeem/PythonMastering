```python
import numpy as np

# 1. Defining a Transformation Matrix (A) and a Vector (x)
# Let A be a 2D transformation that stretches x by 2 and shears y
A = np . array ([[2 , 1] , [0 , 1]])
x = np . array ([1 , 1])

# Matrix - Vector Multiplication ( Applying the transformation )
transformed_x = A @ x # The ’@’ operator denotes matrix multiplication in Python
print ( f" Original x: {x}")
print ( f" Transformed x: { transformed_x }\n")

# 2. Matrix Multiplication ( Composition of Transformations )
# Let B be a 90 - degree rotation matrix
B = np . array ([[0 , -1] , [1 , 0]])

# Applying B then A (A @ B)
C = A @ B
print ( f" Combined Transformation Matrix C:\n{C}\n")

# 3. Solving a System of Linear Equations : Ax = b
# 3 x_1 + 2x_2 = 12
# 1 x_1 - 1x_2 = -1
coeffs = np . array ([[3 , 2] , [1 , -1]])
b = np . array ([12 , -1])

# Method 1: Using the Inverse ( Mathematically elegant , computationally slow for big data )
A_inv = np . linalg . inv ( coeffs )
x_method1 = A_inv @ b
print ( f" Solution via Inverse : { x_method1 }")

# Method 2: Using NumPy ’s solver ( Optimized for performance and numerical stability )
x_method2 = np . linalg . solve ( coeffs , b )
print ( f" Solution via linalg . solve : { x_method2 }")
```
