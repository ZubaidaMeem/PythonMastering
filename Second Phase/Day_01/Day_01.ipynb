```python
import numpy as np

# ==========================================
# Problem 1: Linear Combinations
# ==========================================
h1 = np . array ([1200 , 2 , 15])
h2 = np . array ([1500 , 3 , 10])
h3 = np . array ([800 , 1 , 20])

weights = np . array ([0.5 , 0.3 , 0.2])

# TODO : Compute the linear combination using NumPy operations
# y = None # Replace with your code
y = weights[0] * h1 + weights[1] * h2 + weights[2] * h3

print (" Problem 1 Result :")
print ( f" Weighted House Vector : {y}\n")

# ==========================================
# Problem 2: Detecting Multi - collinearity
# ==========================================
# Feature columns : [ Temp (C), Humidity (%) , Temp (F)]
# Note : Temp (F) = Temp (C) * 1.8 + 32.
# Because of the +32 ( affine shift ), we will add a column of 1s ( bias )
# to accurately check for linear dependence in ML data .

dataset = np . array ([[1 , 20 , 45 , 68] ,
[1 , 25 , 50 , 77] ,
[1 , 30 , 60 , 86] ,
[1 , 15 , 30 , 59]
])

# TODO : Calculate the rank of the dataset matrix
# matrix_rank = None # Replace with your code
matrix_rank = np.linalg.matrix_rank(dataset)

# TODO : Write a conditional statement to check if features are independent
# Hint : Compare the rank to the number of features ( columns ).
# is_independent = None # Replace with your code
is_independent = matrix_rank == dataset.shape[1]

print (" Problem 2 Result :")
print ( f" Matrix Rank : { matrix_rank }")
print ( f" Are features linearly independent ? { is_independent }\n")

# ==========================================
# Problem 3: Standard Basis Generation
# ==========================================
def generate_standard_basis ( n ) :
    """
    Generates the standard basis for R^n.
    Returns a list of n numpy arrays , each of shape (n ,)
    """
    basis_vectors = []
    # TODO : Write a loop or use NumPy built -in functions ( like np.eye )
    # to generate the standard basis vectors e_1 , e_2 , ... , e_n
    x = np.eye(n)
    basis_vectors = [x[i] for i in range(n)]
    return basis_vectors

print (" Problem 3 Result :")
n_dim = 4
basis = generate_standard_basis ( n_dim )
for i , b in enumerate ( basis ) :
    print ( f"e_{i+1} = {b}")
```
